```python
#from flask import Flask,render_template,redirect,url_for,jsonify,request,make_response,session
#from flask_bootstrap import Bootstrap
import pandas as pd
import numpy as np
from numpy import NAN   #用来判断是否有NAN
from pandas import DataFrame
from pymongo import MongoClient
import time,datetime,json
#from flask_httpauth import HTTPBasicAuth
#from flask_cache import Cache
import os
#from flask import send_from_directory
#auth = HTTPBasicAuth()
client=MongoClient('mongodb://root:' + '5768116' + '@121.196.220.14')


db5 = client.piaofen
collection5 = db5.piaofen
cursor = collection5.find({"$and":[{'time':'2019-04-02'}]})
present_time = pd.DataFrame(list(cursor))

```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties  
font_set2 = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=9) #用于在散点图中输出中文
print(present_time)

huatudata3=present_time[['hanglei1','hanglei2','shou','chu','shoudai','chudai','time2']]

huatudata_y=huatudata3[huatudata3['hanglei2']==1]
huatudata_y.rename(columns={'shou':'yshou','chu':'ychu'}, inplace=True)
huatudata_y=huatudata_y[['yshou','ychu']]
huatudata_z=huatudata3[huatudata3['hanglei2']==0]
huatudata_z.rename(columns={'shou':'zshou','chu':'zchu'}, inplace=True)
huatudata_z=huatudata_z[['zshou','zchu']]

huatudata3 = pd.concat([huatudata_y, huatudata_z], axis=1)
huatudata3=huatudata3.fillna(0)
#print(huatudata3)

huatudata2=huatudata3.cumsum(0)   #累计图
huatudata=present_time[['hanglei2','time2']]
result_qun = pd.concat([huatudata2, huatudata], axis=1)
result_qun.to_csv('kkkk.csv')
plt.scatter(y=result_qun['yshou'],x=result_qun['ychu'],marker ='+',label='黄色为银行，紫色为中介',edgecolors='face',s=20)  #制作散点图

plt.scatter(y=result_qun['zshou'],x=result_qun['zchu'],marker ='+',label='黄色为银行，紫色为中介',edgecolors='face',s=20)  #制作散点图

plt.legend(prop=font_set2)  #显示lable
plt.show()


```

```python
plt.scatter(x=result_qun['time2'],y=result_qun['yshou'],marker ='+',label='yshou',edgecolors='face',s=20)  #制作散点图

plt.scatter(x=result_qun['time2'],y=result_qun['ychu'],marker ='+',label='ychu',edgecolors='face',s=20)  #制作散点图
plt.scatter(x=result_qun['time2'],y=result_qun['zshou'],marker ='+',label='zshou',edgecolors='face',s=20)  #制作散点图

plt.scatter(x=result_qun['time2'],y=result_qun['zchu'],marker ='+',label='zchu',edgecolors='face',s=20)  #制作散点图

plt.legend(prop=font_set2)  #显示lable
plt.show()
```

```python
result_quny=result_qun[['yshou','ychu']]
result_quny=result_quny.drop_duplicates(['ychu'],keep='last')
result_quny=result_quny.set_index(["ychu"],drop=True)



result_qunz=result_qun[['zshou','zchu']]
result_qunz=result_qunz.drop_duplicates(['zchu'],keep='last')

result_qunz=result_qunz.set_index(["zchu"],drop=True)
result_qunw = pd.concat([result_quny, result_qunz], axis=1)

#result_qun= result_qun.sort_values(by=index, ascending=True)
#result_qunw.sort_index()
result_qunw.to_csv('ddd.csv')
print(result_qunw.index)
plt.scatter(x=result_qunw.index,y=result_qunw['yshou'],marker ='+',label='yshou',edgecolors='face',s=20)  #制作散点图

plt.scatter(x=result_qunw.index,y=result_qunw['zshou'],marker ='+',label='ychu',edgecolors='face',s=20)  #制作散点图

plt.legend(prop=font_set2)  #显示lable
plt.show()
```
