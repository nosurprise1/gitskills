```python
import pdfplumber
import re
import pandas as pd

dfz = pd.read_table('D:\downloadFiles\yuebao\mu.csv',sep=',',encoding='GB18030')   #用dictreader根据行内容查找

def not_empty(s):
    return s and s.strip()



zongz=pd.DataFrame( {'统计项目':'',
                     '笔数':'',
                       '余额（亿元）': '',
                       '日期':['']
                         })
print(len(dfz))

for i in range(0,len(dfz)):
    #print(dfz.loc[i,'1文件名.txt'])
    if '余额' in dfz.loc[i,'1文件名.txt']  and '2019' not in dfz.loc[i,'1文件名.txt'] :
         path = ('D:\downloadFiles\yuebao')+'\\'+ dfz.loc[i,'1文件名.txt']
         print(path)
         pdf = pdfplumber.open(path)



#        for page in pdf.pages:  读取pdf的第一页
         hang=pdf.pages[0].extract_text().split('\n')
      #   print(page.extract_text())

         hang1=hang[3].split(' ')
         hang1=list(filter(not_empty, hang1))  #列表去除空值
         hang1=hang1[len(hang1)-4:]   #取列表的最后4位

         zong=pd.DataFrame( {hang[2].split(' ')[0]:'',
                   hang[2].split(' ')[1]: hang1[0],
                   hang[2].split(' ')[2]: hang1[1],
                        '日期':[hang[1].split('—')[0]]
                     })

         for i in range(4,len(hang)):

            hang1=hang[i].split(' ')
            hang1=list(filter(not_empty, hang1))  #列表去除空值
            if len(hang1)>3 :
                 hang1=hang1[len(hang1)-4:]   #取列表的最后4位

                 zong1=pd.DataFrame( {hang[2].split(' ')[0]:'',
                            hang[2].split(' ')[1]: hang1[0],
                            hang[2].split(' ')[2]: hang1[1],
                                 '日期':[hang[1].split('—')[0]]
                              })

                 zong=pd.concat([zong,zong1])
         zong= zong.reset_index(drop=True)    #重新定义索引
         zong.loc[0,'统计项目']='银票承兑余额'
         zong.loc[1,'统计项目']='商票承兑余额'
         zong.loc[2,'统计项目']='承兑余额合计'
         zong.loc[3,'统计项目']='银票贴入余额'
         zong.loc[4,'统计项目']='商票贴入余额'
         zong.loc[5,'统计项目']='贴入余额合计'
         zong.loc[6,'统计项目']='回购银票余额'
         zong.loc[7,'统计项目']='回购商票余额'
         zong.loc[8,'统计项目']='回购余额合计'
                
         print(zong)
    
     
         zongz=pd.concat([zong,zongz])
         zongz.to_csv('zongz-yu.csv')                
                        
                        
                        

        
```
