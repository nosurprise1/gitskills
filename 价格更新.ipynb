# 信贷评级更新 

```python
import csv,xlrd,xlwt,datetime,time
import pandas as pd





riqi=input('请输入要最新授信表的日期:')
url=input('请输入要最新授信表的文件名（包括后缀）:')

shouxin=pd.read_excel('D:/20190409部分/20180929桌面备份/报买断业务/转贴现买断价格计算/评级更新/%s'%url,sheetname=1,encoding='GB18030')


shouxin=shouxin[['客户名称','客户状态','额度到期日','本年度信用评级']]

shouxin['d']=0
am=['A-','A','A+','AA-','AA','AA+','AAA-','AAA','AAA+']
bbb=['BBB+']

shouxin= shouxin.reset_index(drop=True)    #重新定义索引


#一、导入指导价
jia_df=pd.read_excel('D:/20190409部分/20180929桌面备份/报买断业务/转贴现买断价格计算/jia.xls',encoding='GB18030')
#jia_df=pd.read_table('jia2.csv', sep=',',encoding='GB18030')

shouxin['客户名称']=shouxin['客户名称'].astype(str)


for i in range(0,len(shouxin)):
  if '银行' in shouxin.loc[i,'客户名称']:
        shouxin.loc[i,'d']=4  #挑选出所有银行

  if '含' in  str(shouxin.loc[i,'额度到期日']):
     shouxin.loc[i,'额度到期日']='2010-01-01'  #把shouxin的表中有不规范的到期日格式排除

  shouxin.loc[i,'额度到期日']=str(shouxin.loc[i,'额度到期日']).replace(' 00:00:00','')



for j in range(0,len(shouxin)):
  for i in range(0,len(jia_df)):
    if jia_df.loc[i,'银行类型'] in shouxin.loc[j,'客户名称']:
        shouxin.loc[j,'d']='已有'
        if jia_df.loc[i,'授信有效']==shouxin.loc[j,'客户状态'] and jia_df.loc[i,'我行评级']==shouxin.loc[j,'本年度信用评级'] and jia_df.loc[i,'生效截至日']==shouxin.loc[j,'额度到期日']  :
            continue
        else:


            print(jia_df.loc[i,'银行类型'] )
            print(jia_df.loc[i,'授信有效'] )
            print(shouxin.loc[j,'客户状态'] )
            print(jia_df.loc[i,'我行评级'])
            print(shouxin.loc[j,'本年度信用评级'])


            
            jia_df.loc[i,'更新日期']=riqi
            jia_df.loc[i,'授信有效']=shouxin.loc[j,'客户状态']
            jia_df.loc[i,'我行评级']=shouxin.loc[j,'本年度信用评级']
            jia_df.loc[i,'生效截至日']=shouxin.loc[j,'额度到期日'] 
        break
  


shouxin=shouxin[shouxin['d']!='已有']
shouxin= shouxin.reset_index(drop=True)    #重新定义索引
for i in range(0,len(shouxin)):
    if shouxin.loc[i,'本年度信用评级'] in am and ('银行' in shouxin.loc[i,'客户名称']):
        shouxin.loc[i,'d']=2      #挑选出A-及以上的银行
    
    elif shouxin.loc[i,'本年度信用评级'] in bbb and ('银行' in shouxin.loc[i,'客户名称']):
        shouxin.loc[i,'d']=3      #挑选出BBB+及以上的银行

    elif ('财务' in shouxin.loc[i,'客户名称']):
        shouxin.loc[i,'d']=5      #财务公司        

jia_df=jia_df.set_index('银行类型')    
url3='D:/20190409部分/20180929桌面备份/报买断业务/转贴现买断价格计算/xin.xls'
writer = pd.ExcelWriter(url3)
shouxin.to_excel(writer,'Sheet1')
writer.save()


url3='D:/20190409部分/20180929桌面备份/报买断业务/转贴现买断价格计算/jia-评级.xls'
writer = pd.ExcelWriter(url3)
jia_df.to_excel(writer,'Sheet1')
writer.save()

```
