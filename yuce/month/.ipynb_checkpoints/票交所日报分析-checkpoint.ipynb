```python

for i in range(0,len(df2)):
    highpoints = re.compile(u'[\U00010000-\U0010ffff]')
    nickname = highpoints.sub(u'',df2.loc[i,'nickname'])

    df2.loc[i,'nickname']= nickname
    for j in range(0,len(number)):
        
        if str(df2.loc[i,'nickname']) in str(number.loc[j,'姓名']):
            df2.loc[i,'sheng']=number.loc[j,'省份']
            df2.loc[i,'银行']=number.loc[j,'银行分类']
            break




df2=df2[(df2['sheng']!='\\')& (df2['sheng']!='/') & (df2['sheng']!='总行')]


df_b=df2.groupby(df2['银行']).sum()
df_b=df_b[['shou','chu']]
df_b.rename(columns={'shou':'收票', 'chu':'出票'}, inplace = True)





names=df_b.index.tolist()
x = range(len(names))
df_b.plot(kind='bar', alpha=0.5,stacked=False,label='')
plt.xticks(x,names,rotation = 45,fontproperties=font_set)
plt.xlabel('')

plt.legend(prop=font_set,loc='upper left')  #显示lable
plt.ylabel('广告数',fontproperties=font_set)
plt.title('3月微信广告银行类别分析', fontproperties=font_set) 

fig =plt.gcf()
fig.set_size_inches(18, 12)
fig.savefig('3月微信广告银行分类.png')

fig.show()






df_a=df2.groupby(df2['sheng']).sum()
df_a=df_a[['shou','chu']]
df_a.rename(columns={'shou':'收票', 'chu':'出票'}, inplace = True)

names=df_a.index.tolist()
print(df_a)
x = range(len(names))
df_a.plot(kind='bar', alpha=0.5,stacked=False,label='')
plt.xticks(x,names,rotation = 45,fontproperties=font_set)
plt.xlabel('')

plt.legend(prop=font_set,loc='upper left')  #显示lable
plt.ylabel('广告数',fontproperties=font_set)
plt.title('3月微信广告地域分析', fontproperties=font_set) 

fig =plt.gcf()
fig.set_size_inches(18, 9)
fig.savefig('3月微信广告省份分析.png')

fig.show()

        

df_c=df2.groupby(df2['time']).sum()
df_c=df_c[['shou','chu']]
df_c['收票/出票']=df_c['shou']/df_c['chu']
cy1=df_c['收票/出票'].tolist()
df_c=df_c[['shou','chu']]

df_c.rename(columns={'shou':'收票', 'chu':'出票'}, inplace = True)
names=df_c.index.tolist()
print(df_c)
x = range(len(names))
#names=df_a.index.tolist()
#print(names)
#x = range(len(names))
df_c.plot(kind='bar', alpha=0.5,stacked=False,label='')
plt.xticks(x,names,rotation = 45,fontproperties=font_set)
plt.xlabel('')
plt.ylabel('广告数',fontproperties=font_set)
plt.legend(prop=font_set,loc='upper left')  #显示lable

plt2=plt.twinx()
plt2.plot(cy1,alpha=0.7)
plt2.set_ylabel('收票/出票',fontproperties=font_set)
plt2.legend(prop=font_set,loc='upper right')  #显示lable

plt.title('3月微信广告时间序列分析', fontproperties=font_set) 

fig =plt.gcf()
fig.set_size_inches(18, 9)
fig.savefig('3月微信广告时间序列分析.png')

fig.show()
```
