# 分析变动

```python
import pandas as pd
import numpy,math,time,datetime
jin=pd.read_excel('jin.xls',encoding='GB18030')
shijian0=time.strftime('%Y-%m-%d',time.localtime(time.time()))
shijian=datetime.datetime.strptime(shijian0 , "%Y-%m-%d")#转为日期格式
shijian=str(shijian.strftime("%Y-%m-%d"))
jincolumn=jin.columns.values.tolist()  
jincolumn.remove('客户名称')
jincolumn.remove('银行分类')

jincolumn.remove('生效起始日')
jincolumn.remove('生效截至日')
print('已经收集的日期有：')
print(jincolumn)
zuijinri=max(jincolumn)
jincolumn.remove(zuijinri)
shangyiri=max(jincolumn)



print('\n')
print('最近日期：')

print(zuijinri)
print('上一日期：')
print(shangyiri)


jin[shangyiri]=jin[shangyiri].astype('float')/100000000
jin[zuijinri]=jin[zuijinri].astype('float')/100000000
jin['改变']=0
for i in range(0,len(jin)):
    if math.isnan(jin.loc[i,zuijinri]) and  not math.isnan(jin.loc[i,shangyiri]):
             print('删除名单')
             print('%s     %s     %s    %s'%(jin.loc[i,'客户名称'],jin.loc[i,shangyiri],jin.loc[i,'生效起始日'],jin.loc[i,'生效截至日']))
             jin.loc[i,'改变']=1
    elif not math.isnan(jin.loc[i,zuijinri]) and  math.isnan(jin.loc[i,shangyiri]):    
             print('新增名单:')
             print('%s     %s     %s    %s'%(jin.loc[i,'客户名称'],jin.loc[i,zuijinri],jin.loc[i,'生效起始日'],jin.loc[i,'生效截至日']))
             jin.loc[i,'改变']=1
    jin.loc[i,'%s变动'%zuijinri]=(jin.loc[i,zuijinri]-jin.loc[i,shangyiri])


jin=jin[jin['改变']!=1]


jin = jin.sort_values(by=('%s变动'%zuijinri), ascending=False)





jin=jin.round({'%s变动'%zuijinri:2})
jin=jin.round({zuijinri:2})
jin=jin.round({shangyiri:2})

jin=jin.reset_index(drop = True)

jin0=jin[['客户名称','生效截至日',zuijinri,'%s变动'%zuijinri]]
jin0=jin0.reset_index(drop = True)
#print(jin0.head(15))
#print('\n')
#print(jin0.tail(15))

print('\n')

#二、各银行分类的可用余额变动

jin['生效截至日']=jin['生效截至日'].astype('str')
jin=jin[jin['生效截至日']!='']
for i in range(0,len(jin)):
   # print(jin.loc[i,'生效截至日'])
   # print(jin.loc[i,'生效截至日'])
    jin.loc[i,'生效截至日']=jin.loc[i,'生效截至日'].replace('.0','')
    
    
jin['生效截至日']=pd.to_datetime(jin['生效截至日'])
jin1=jin[jin['生效截至日']>=shijian]
jin1=jin1.reset_index(drop = True)
jin1 = jin1.sort_values(by=('%s'%zuijinri), ascending=True)
jin1=jin1.reset_index(drop = True)
jin1=jin1[['银行分类',zuijinri,'%s变动'%zuijinri,'客户名称']]
jinfen1=jin1.groupby([jin1['银行分类']]).sum()
jinfen2=jin1.groupby([jin1['银行分类'],jin1['客户名称']]).sum()

jinfen1=jinfen1.reset_index(drop = False)
for i in range(0,len(jinfen1)):
    jinfen1.loc[i,'银行分类']= jinfen1.loc[i,'银行分类'].replace('（国股贴）','')
jinfen1=jinfen1.set_index('银行分类')  
print('一、可用余额总共新增：%s亿元，剔除过期授信后，可用余额总共新增：%s亿元。'%(jin['%s变动'%zuijinri].sum().round(2),jinfen1['%s变动'%zuijinri].sum().round(2)))
print('\n')

print('二、各类银行的可用余额变动汇总：')
print(jinfen1.round(2))





#三、各银行分类的可用余额变动
print('\n')
print('三、各类银行的可用余额变动明细：')

jinfen2=jinfen2.reset_index(drop = False)
jinfen2_guo=jinfen2[jinfen2['银行分类']=='国股']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('国股类银行的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')

jinfen2_guo=jinfen2[jinfen2['银行分类']=='大商']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('大商类银行的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')

jinfen2_guo=jinfen2[jinfen2['银行分类']=='一类准入']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('一类准入的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')
jinfen2_guo=jinfen2[jinfen2['银行分类']=='二类准入']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('二类准入的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')

jinfen2_guo=jinfen2[jinfen2['银行分类']=='其他授信银行']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('其他授信银行的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')
jinfen2_guo=jinfen2[jinfen2['银行分类']=='准入财务公司（国股贴）']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('准入财务公司的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
print('\n')
jinfen2_guo=jinfen2[jinfen2['银行分类']=='商票、授信财务公司（国股贴）']
jinfen2_guo = jinfen2_guo.sort_values(by=('%s变动'%zuijinri), ascending=True)
print('商票、授信财务公司的可用余额变动明细：')
jinfen2_guo=jinfen2_guo.reset_index(drop = True)
print(jinfen2_guo)
```
