```python
import csv,xlrd,xlwt,datetime,time,re
import numpy as np

import pandas as pd


#一、导入指导价

#二、打开对方清单
def suanqingdan():
    global zongfen
    df=pd.read_excel('计算分数/20190627日终余额明细.xls',encoding='GB18030')
    df['承兑人'] = (df['承兑人']).astype('str') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式
    df['票面金额'] = (df['票面金额']).astype('float') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式

        
        
    df['筛选']=''    
    for i in range(0,len(df)):
        if '锦州银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='锦州银行'
        elif '渤海银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='渤海银行'
        elif '哈尔滨银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='哈尔滨银行'
        elif '辽阳银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='辽阳银行'
        elif '盛京银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='盛京银行'
        elif '阜新银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='阜新银行'      
        elif '营口银行' in df.loc[i,'承兑人']:
             df.loc[i,'筛选']='营口银行'   
    df=df[df['筛选']!='']
    df['协议编号'] = (df['协议编号']).astype('str') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式
    df['到期日'] = (df['到期日']).astype('str') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式
    df['付款行行号'] = (df['付款行行号']).astype('str') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式
    df['交易日期'] = (df['交易日期']).astype('str') #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式

    df.to_excel('挑选票据清单.xls')

                
    df=df[['票号','票面金额','出票日','到期日','承兑人','出票人','归属机构','交易对手','筛选','协议编号']]
    


    huizong=df['票面金额'].sum()
   # huizong= huizong.reset_index(drop=False)    #重新定义索引
  #  huizong['票面金额']=huizong['票面金额']/100000000
    print(huizong)

    
    huizong=df['票面金额'].groupby([df['筛选']]).sum()
    huizong= huizong.reset_index(drop=False)    #重新定义索引
    huizong['票面金额']=huizong['票面金额']/100000000
    print(huizong)

    
    huizong1=df.drop_duplicates(['协议编号'],keep='first')   #删除连接中的重复行

    huizong1=huizong1['协议编号'].groupby([huizong1['归属机构'],huizong1['筛选'],huizong1['交易对手']]).size()
    huizong1= huizong1.reset_index(drop=False)    #重新定义索引
    print(huizong1)
    
    huizong2=df['票面金额'].groupby([df['交易对手']]).sum()
    huizong2= huizong2.reset_index(drop=False)    #重新定义索引
    huizong2.to_csv('uuu.csv')
    print(huizong2)   
    
    
suanqingdan()

```
