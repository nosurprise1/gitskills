```python
from pymongo import MongoClient
import cgi
import re,datetime,json,time
from math import  floor
import csv,xlrd,xlwt
import pdfplumber

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties  
client=MongoClient('mongodb://root:' + '5768116' + '@121.196.220.14')
font_set = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=12) #用于在散点图中输出中文
font_set2 = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=9) #用于在散点图中输出中文
import matplotlib as mpl   #显示中文


import statsmodels.api as sm
import numpy as np
import pandas as pd
db = client.piaojiaosuo4 
collectionl = db.piaojiaosuo4  
cursor = collectionl.find()
df= pd.DataFrame(list(cursor))





def not_empty(s):
    return s and s.strip()



def shangchuan():
    zhiling=input('    请输入指令“1”来确认上传余额：')
    if zhiling!='1':
           print('    终止。')
           return

    zhiling2=input('    请输入文件名pdf：')
    if zhiling2=='':
           print('    终止。')
           return    

    url2=('D:\downloadFiles\yuebao')+'\\'+ zhiling2

    pdf2 = pdfplumber.open(url2)




#        for page in pdf.pages:  读取pdf的第一页
    hang=pdf2.pages[0].extract_text().split('\n')

    hang1=hang[3].split(' ')
    hang1=list(filter(not_empty, hang1))  #列表去除空值
    hang1=hang1[len(hang1)-4:]   #取列表的最后4位

    zong=pd.DataFrame( {
                 })
    print(len(hang))
    for i in range(3,len(hang)):

        hang1=hang[i].split(' ')
        hang1=list(filter(not_empty, hang1))  #列表去除空值
        if len(hang1)>3 :
             hang1=hang1[len(hang1)-4:]   #取列表的最后4位

             zong1=pd.DataFrame( {hang[2].split(' ')[0]:'',
                        '余额（亿元）': float(hang1[1].replace(',','')),
                       # hang[2].split(' ')[2]: hang1[1],
                             '日期':[hang[1].split('日')[0]]
                          })
             print(zong1)
             zong=pd.concat([zong,zong1])
    zong= zong.reset_index(drop=True)    #重新定义索引
    zong.loc[0,'统计项目']='银票承兑余额'
    zong.loc[1,'统计项目']='商票承兑余额'
    zong.loc[2,'统计项目']='承兑余额合计'
    zong.loc[3,'统计项目']='银票贴入余额'
    zong.loc[4,'统计项目']='商票贴入余额'
    zong.loc[5,'统计项目']='贴入余额合计'

    zong= zong.reset_index(drop=True)    #重新定义索引
    for i in range(0,len(zong)):
        zong.loc[i,'日期']=zong.loc[i,'日期'].replace('年','/').replace('月','/').replace('日','')

    zong['日期'] = pd.to_datetime(zong['日期'])  #,format='%Y%m%d'  吧yyyy-m-d 转化为yyyy-mm-dd的时间格式
    zong['日期'] = zong['日期'].dt.strftime('%Y/%m/%d').astype('str')

    print(zong)
    stop
    #print collection
    records = json.loads(zong.T.to_json()).values()
    collectionl.insert(records)

shangchuan()












df= df.reset_index(drop=True)    #重新定义索引
df['日期']=df['日期'].astype('str')
df['余额（亿元）']=df['余额（亿元）'].astype('str')
df=df.set_index(['日期'])




df1=df[df['统计项目']=='银票承兑余额']
df1=df1[['余额（亿元）']]
df1.rename(columns={'余额（亿元）':'银票承兑余额'}, inplace=True)
df5=df[df['统计项目']=='商票承兑余额']
df5=df5[['余额（亿元）']]
df5.rename(columns={'余额（亿元）':'商票承兑余额'}, inplace=True)

df2=df[df['统计项目']=='银票贴入余额']
df2=df2[['余额（亿元）']]
df2.rename(columns={'余额（亿元）':'银票贴入余额'}, inplace=True)
df6=df[df['统计项目']=='商票贴入余额']
df6=df6[['余额（亿元）']]

df6.rename(columns={'余额（亿元）':'商票贴入余额'}, inplace=True)

df3=df[df['统计项目']=='回购银票余额']
df3=df3[['余额（亿元）']]

df3.rename(columns={'余额（亿元）':'银票回购余额'}, inplace=True)
df4=df[df['统计项目']=='回购商票余额']
df4=df4[['余额（亿元）']]

df4.rename(columns={'余额（亿元）':'商票回购余额'}, inplace=True)



df=pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
#df2=df1[['余额（亿元）']]


df=df[['银票承兑余额','商票承兑余额','银票贴入余额','商票贴入余额','银票回购余额',
    '商票回购余额']]
df= df.reset_index(drop=False)    #重新定义索引
print(df)
```
