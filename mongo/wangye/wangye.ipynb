# quxian（个人上报的利率记录）数据库

```python
# -*- coding: cp936 -*-
import datetime
import json,csv
import pandas as pd

from pymongo import MongoClient
client=MongoClient('mongodb://root:' + '5768116' + '@121.196.220.14')
db = client.quxian          #得到数据库
collection = db.quxian      #得到数据集合

#cursor = collection.find()
cursor = collection.find({"$or":[{'来源':{'$regex':'18957170906'}},{'共享':'共享'}]})

quxian_df=pd.DataFrame(list(cursor))
quxian_dfz=quxian_df[quxian_df['日期']==quxian_df.loc[(len(quxian_df)-1),'日期']]  #选择最后一天的数据

#    把利率曲线中邮储和农行的价格始终放上去
quxian_dfsz1=quxian_df[quxian_df['报价行'].isin(['邮储'])]
quxian_dfsz1= quxian_dfsz1.reset_index(drop=False)    #重新定义索引
quxian_dfsz1=quxian_dfsz1[quxian_dfsz1['日期']==quxian_dfsz1.loc[(len(quxian_dfsz1)-1),'日期']]  #选择最后一天的数据

     #选择最后一天的数据
quxian_dfsz2=quxian_df[quxian_df['报价行'].isin(['农行'])]  #选择最后一天的数据
quxian_dfsz2= quxian_dfsz2.reset_index(drop=False)    #重新定义索引
quxian_dfsz2=quxian_dfsz2[quxian_dfsz2['日期']==quxian_dfsz2.loc[(len(quxian_dfsz2)-1),'日期']]  #选择最后一天的数据


quxian_df = pd.concat([quxian_dfsz1, quxian_dfsz2,quxian_dfz], axis=0)
quxian_df =quxian_df .drop_duplicates('_id', keep='last')  #删除重复

print(quxian_df)


```
